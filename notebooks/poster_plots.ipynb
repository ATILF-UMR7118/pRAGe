```python
def z_norm(column):
    return (column - column.mean()) / column.std()
```

```python
df= pd.DataFrame(vals, columns=['bert', 'bert-std','bleurt','bleurt-std','bleu','bleu-std', 'rouge', 'rouge-std'])
divider=6
df = df.drop([3, 5, 9, 11]); divider=4 # to remove prage drbert config
df
```

```python
df_z_norm = df.apply(z_norm)
df_z_norm

plt.scatter(list(range(len(df[:divider]))), df_z_norm.bert.values[:divider], marker='p', label='bertscore', color='blue', alpha=0.35, s=150)
plt.scatter(list(range(len(df[:divider]))), df_z_norm.bleurt.values[:divider], marker='s', label='bleurt', color='olive', alpha=0.35, s=150)
plt.scatter(list(range(len(df[:divider]))), df_z_norm.bleu.values[:divider], marker='o', label='bleu', color='purple', alpha=0.35, s=150)
plt.scatter(list(range(len(df[:divider]))), df_z_norm.rouge.values[:divider], marker='^', label='rouge', color='chocolate', alpha=0.35, s=150)

plt.scatter(list(range(len(df[divider:]))), df_z_norm.bert.values[divider:], marker='p',color='blue',label='bertscore-FT', s=150)
plt.scatter(list(range(len(df[divider:]))), df_z_norm.bleurt.values[divider:], marker='s', color='olive',label='bleurt-FT', s=150)
plt.scatter(list(range(len(df[divider:]))), df_z_norm.bleu.values[divider:], marker='o', color='purple',label='bleu-FT', s=150)
plt.scatter(list(range(len(df[divider:]))), df_z_norm.rouge.values[divider:], marker='^', color='chocolate',label='rouge-FT', s=150)


if divider == 6:
    plt.xticks(list(range(0,divider)), labels=['BARTHEZ','BIOMISTRAL','CAMEMBERT+BARTHEZ','dRBERTbARTHEZ', 'CAMEMBERT+BIOMISTRAL','dRbertBIOM'], rotation=25)
else:
    plt.xticks(list(range(0,divider)), labels=['BARTHEZ','BIOMISTRAL','CAMEMBERT+BARTHEZ','CAMEMBERT+BIOMISTRAL'], rotation=7)

plt.legend(ncols=2, loc=(.39,0.75))
plt.tight_layout()
plt.savefig('autoeval.pdf')
```
